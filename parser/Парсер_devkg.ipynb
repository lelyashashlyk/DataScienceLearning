```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### –ü–æ–ª—É—á–∏–º —Å—Å—ã–ª–∫–∏ –≤–∞–∫–∞–Ω—Å–∏–π —Å–æ –≤—Å–µ—Ö —Å—Ç—Ä–∞–Ω–∏—Ü, —Å –ø–æ–º–æ—â—å—é –∏—Ç–µ—Ä–∞—Ü–∏–∏.
–°—Å—ã–ª–∫–∏ –æ—Ç–±–∏—Ä–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ —É –∞–∫—Ç—É–∞–ª—å–Ω—ã—Ö –≤–∞–∫–∞–Ω—Å–∏–π –∫–ª–∞—Å—Å–∞ "link". –ù–µ–∞–∫—Ç—É–∞–ª—å–Ω—ã–µ –≤–∞–∫–∞–Ω—Å–∏–∏ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∞—Ç –∫ –¥—Ä—É–≥–æ–º—É –∫–ª–∞—Å—Å—É

```python
url = 'https://devkg.com/ru/jobs?page='
urls = []
for i in range(1, 13):
  url_i = url+str(i)
  response = requests.get(url_i)
  soup = BeautifulSoup(response.text, 'html.parser')
  # for j in soup.find_all('a', class_="link"):
  #   urls.append(j['href'])

  for tag in soup.select('a.link:not(.archived)'): # –≤—ã–±–∏—Ä–∞–µ–º –ø–æ CSS-—Å–µ–ª–µ–∫—Ç–æ—Ä—É, —Ç–æ–ª—å–∫–æ —Å—Å—ã–ª–∫–∏, –≤–Ω—É—Ç—Ä–∏ –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ—Ç —Å—Å—ã–ª–æ–∫ –∫–ª–∞—Å—Å–∞ link archived
    urls.append(tag['href'])

urls
```

### –ü—Ä–æ–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –ø–æ–ª—É—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –æ –ø–æ–∑–∏—Ü–∏–∏
–û–Ω–∞ –±—É–¥–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å –ø–æ–∑–∏—Ü–∏—é, –∑–ø, –∫–æ–º–ø–∞–Ω–∏—é, —Ç–∏–ø (–æ—Ñ–∏—Å\—É–¥–∞–ª–µ–Ω–∫–∞) –∏ –æ–ø–∏—Å–∞–Ω–∏–µ —Å–∞–º–æ–π –≤–∞–∫–∞–Ω—Å–∏–∏

```python
def get_jobs_from_link(soup):
  job = {}
  job['position'] = soup.find('h1').text
  job['salary'] = soup.find('div', class_ = 'price').text.split('\n')[1].strip()
  if type(soup.find('div', class_ ="company")) is type(None):
    job['company'] = soup.find('div', class_ ="organizations-item-field name").text.split()[1]
  else:
    job['company'] = soup.find('div', class_ ="company").text.split()[1]
  job['type'] = ' '.join(soup.find('div', class_ ="type").text.split()[1:])
  job['description'] = ' '.join(soup.find('div', class_ ="text").text.split()[2:])
  return job
```

### –ü–æ–ª—É—á–∏–º –¥–∞–Ω–Ω—ã–µ –∏—Å–ø–æ–ª—å–∑—É—è –Ω–∞—à—É —Ñ—É–Ω–∫—Ü–∏—é

```python
main_jobs_list = []
for url in urls:
  response = requests.get('https://devkg.com' + url)
  soup = BeautifulSoup(response.text, 'html.parser')
  print('https://devkg.com' + url)
  main_jobs_list.append(get_jobs_from_link(soup))
```

## –ù–µ–º–Ω–æ–≥–æ –Ω–µ—É–¥–æ–±–Ω—ã–π –∫–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∫–∞ –Ω–µ —É–¥–∞–ª—è—é

–ù–∏–∂–µ –¥–∞–Ω–Ω—ã–µ –ø–æ–ª—É—á–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ —Å –æ–¥–Ω–æ–π —Å—Ç—Ä–∞–Ω–∏—Ü—ã –∏–∑ —Å–ø–∏—Å–∫–∞ –≤–∞–∫–∞–Ω—Å–∏–π. –ò–∑–≤–ª–µ—á–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –¥–∞–ª–µ–µ –ø–æ–º–µ—â–∞—é—Ç—Å—è –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –¥–ª—è –±–æ–ª–µ–µ —É–¥–æ–±–Ω–æ–≥–æ –ø—Ä–æ—Å–º–æ—Ç—Ä–∞ –∏—Ç–æ–≥–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö

```python
# –≤—Å–µ —á—Ç–æ –Ω–∏–∂–µ, –æ—Ñ–æ—Ä–º–ª–µ–Ω–Ω–æ–µ –≤ —Ñ—É–Ω–∫—Ü–∏—é

def get_jobs_from_list(soup):
  jobs_div = [div for a in soup.find_all('a', class_ = 'link') for div in a.find_all('div', class_='information information-main')]
  jobs_list = []
  salary_list = []
  type_list = []
  company_list = []
  for i in jobs_div:
    jobs_list.append(i.find('div', class_ ="jobs-item-field position").text.split('\n')[1].strip())
    salary_list.append(i.find('div', class_ ="jobs-item-field price").text.split('\n')[1].strip())
    company_list.append(i.find('div', class_ ="jobs-item-field company").text.split('\n')[1].strip())
    type_list.append(i.find('div', class_ ="jobs-item-field type").text.split('\n')[1].strip())

  return (jobs_list, salary_list, type_list, company_list)
```

–ù–∞—á–∞–ª–æ –ø—Ä–æ–≤–µ—Ä–µ–Ω–Ω–æ–π —á–∞—Å—Ç–∏ –∫–æ–¥–∞

```python
response = requests.get('https://devkg.com/ru/jobs')
soup = BeautifulSoup(response.text, 'html.parser')
jobs_div = [div for a in soup.find_all('a', class_ = 'link') for div in a.find_all('div', class_='information information-main')]
```

```python
jobs_list = []
for i in jobs_div:
  jobs_list.append(i.find('div', class_ ="jobs-item-field position").text.split('\n')[1].strip())
```

```python
salary_list = []
for i in jobs_div:
  salary_list.append(i.find('div', class_ ="jobs-item-field price").text.split('\n')[1].strip())
```

```python
company_list = []
for i in jobs_div:
  company_list.append(i.find('div', class_ ="jobs-item-field company").text.split('\n')[1].strip())
```

```python
type_list = []
for i in jobs_div:
  type_list.append(i.find('div', class_ ="jobs-item-field type").text.split('\n')[1].strip())
```

```python
title = ['position', 'company', 'type', 'salary']
label = [jobs_list, company_list, type_list, salary_list]
fin_jobs = [[j[i] for j in label] for i in range(len(jobs_list))]
```

```python
pd.DataFrame(fin_jobs, columns=title)
```
