```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### Получим ссылки вакансий со всех страниц, с помощью итерации.
Ссылки отбираются только у актуальных вакансий класса "link". Неактуальные вакансии принадлежат к другому классу

```python
url = 'https://devkg.com/ru/jobs?page='
urls = []
for i in range(1, 13):
  url_i = url+str(i)
  response = requests.get(url_i)
  soup = BeautifulSoup(response.text, 'html.parser')
  # for j in soup.find_all('a', class_="link"):
  #   urls.append(j['href'])

  for tag in soup.select('a.link:not(.archived)'): # выбираем по CSS-селектору, только ссылки, внутри которых нет ссылок класса link archived
    urls.append(tag['href'])

urls
```

### Пропишем функцию получения данных о позиции
Она будет содержать позицию, зп, компанию, тип (офис\удаленка) и описание самой вакансии

```python
def get_jobs_from_link(soup):
  job = {}
  job['position'] = soup.find('h1').text
  job['salary'] = soup.find('div', class_ = 'price').text.split('\n')[1].strip()
  if type(soup.find('div', class_ ="company")) is type(None):
    job['company'] = soup.find('div', class_ ="organizations-item-field name").text.split()[1]
  else:
    job['company'] = soup.find('div', class_ ="company").text.split()[1]
  job['type'] = ' '.join(soup.find('div', class_ ="type").text.split()[1:])
  job['description'] = ' '.join(soup.find('div', class_ ="text").text.split()[2:])
  return job
```

### Получим данные используя нашу функцию

```python
main_jobs_list = []
for url in urls:
  response = requests.get('https://devkg.com' + url)
  soup = BeautifulSoup(response.text, 'html.parser')
  print('https://devkg.com' + url)
  main_jobs_list.append(get_jobs_from_link(soup))
```

## Немного неудобный код, который пока не удаляю

Ниже данные получаются только с одной страницы из списка вакансий. Извлеченные данные далее помещаются в датафрейм для более удобного просмотра итоговых данных

```python
# все что ниже, оформленное в функцию

def get_jobs_from_list(soup):
  jobs_div = [div for a in soup.find_all('a', class_ = 'link') for div in a.find_all('div', class_='information information-main')]
  jobs_list = []
  salary_list = []
  type_list = []
  company_list = []
  for i in jobs_div:
    jobs_list.append(i.find('div', class_ ="jobs-item-field position").text.split('\n')[1].strip())
    salary_list.append(i.find('div', class_ ="jobs-item-field price").text.split('\n')[1].strip())
    company_list.append(i.find('div', class_ ="jobs-item-field company").text.split('\n')[1].strip())
    type_list.append(i.find('div', class_ ="jobs-item-field type").text.split('\n')[1].strip())

  return (jobs_list, salary_list, type_list, company_list)
```

```python
response = requests.get('https://devkg.com/ru/jobs')
soup = BeautifulSoup(response.text, 'html.parser')
jobs_div = [div for a in soup.find_all('a', class_ = 'link') for div in a.find_all('div', class_='information information-main')]
```

```python
jobs_list = []
for i in jobs_div:
  jobs_list.append(i.find('div', class_ ="jobs-item-field position").text.split('\n')[1].strip())
```

```python
title = ['position', 'company', 'type', 'salary']
label = [jobs_list, company_list, type_list, salary_list]
fin_jobs = [[j[i] for j in label] for i in range(len(jobs_list))]
```
