# Task 9


В распоряжении начальника имеется бригада рабочих в составе 24 человек. Их нужно распределить на день на два объекта. Если на первом объекте работает t человек, то их суточная зарплата составляет $4t^2$ USD. Если на втором объекте работает t человек, то их суточная зарплата составляет $t^2$ USD. Как нужно распределить на эти объекты бригаду рабочих, чтобы выплаты на их суточную зарплату оказались наименьшими? Сколько USD в этом случае придется заплатить рабочим?



---



---



---




Для решения данной задачи за  $t$  возмем количество работников на первом объекте.

Тогда на втором будет работать $(24-t)$ человек.

Расчет общей стоимости оплаты труда для всех рабочих будет расчитываться следущим образом:

$$s = 4t^2 + (24-t)^2$$

Расчитайте минимально возможное $s$.

Вам необходимо будет перебрать циклом все возможные $t$ от 1 до 24.

(По сути, это задача нахождения минимума функции. Не забывайте функцию min())

````python
task9 = [i for i in range(1, 25)]
posible_s = [4 * i**2 + (24 - i)**2 for i in task9]
print('Все возможные выплаты: ', posible_s)
print('Минимальная выплата: ', min(posible_s))
t1 = posible_s.index(min(posible_s)) # вычисляем индекс минимального значения списка, для нахождения t
t2 = 24 - t1
print('Количество рабочих на первом объекте: ', t1)
print('Количество рабочих на втором объекте: ', t2)
````

**Ход решения**

Представим расстояние между велосипедистами как гипотенузу прямого треугольника. Тогда наименьшая гипотенуза по теореме Пифагора будет равняться квадрату сумм расстояний до перекрестка и пройденных расстояний до момента достижения наименьшего расстояния между велосипедистами за некоторое время соответственно. То есть:

    Гипотенуза min = (S_1 + (40 км/ч * t))**2 + (S_2 + (30 км/ч * t))**2

Прописав функцию нахождения минимальной гипотенузы, найдем ее перебрав значения циклом. В цикле параметром функции будем передавать каждые 3.6 секунд (1ч = 3600 сек -> 1ч * 0.001 = 3600 * 0.001 = 3.6 сек.)


````python
def hypo(t):
   return (5 - 40*t)**2 + (3 - 30*t)**2

min_t = 0
min_hypo = hypo(0)

for i in range(1, 501):
    t = i * 0.001
    h = hypo(t)
    if h < min_hypo:
        min_hypo = h
        min_t = t

print(f"Минимальное расстояние через {round(min_t*60, 2)} минут")
print(f"Минимальное расстояние: {round(min_hypo**0.5, 2)} км")
````
